## Sklearn
Scikit-learn is an open-source machine learning library for Python that provides simple and efficient tools for data analysis and modeling tasks.

With sklearn, you can perform a wide range of machine learning tasks such as classification, regression, clustering, dimensionality reduction, and model evaluation, all through a consistent and easy-to-use Python API.

But first, lets discuss the basic procedure of classification in machine learning. 
1) Get the data
2) Separate the data in features + class
3) Split the data between training, validation, test 
4) fit the model with the train set
5) check training with validation set  
6) predict with the test set 

Sklearn serves as the go-to foundation for establishing baselines in machine learning projects, providing a reliable point of comparison for evaluating cutting-edge models.

Let's see a simple linear regression model

```python
from sklearn.linear_model import LinearRegression

# Generate synthetic data
X = 2 * np.random.rand(100, 1)  # Feature (input)
y = 4 + 3 * X + np.random.randn(100, 1)  # Target (output)

plt.figure(figsize=(8, 3))
plt.scatter(X,y)
```

```python
# Create a linear regression model
model = LinearRegression()
# Fit the model
model.fit(X, y)
```

```python
# Make predictions
X_new = np.linspace(0, 2, 100).reshape(-1, 1)
y_pred = model.predict(X_new)
```

```python
# Plotting
plt.figure(figsize=(8, 3))
plt.scatter(X, y, color='blue', label='Actual Data Points')
plt.plot(X_new, y_pred, color='green', label='Fitted Line')
plt.legend()
plt.show()
```

For some models like linear models, the fitted coefficients and intercepts are stored in these attributes after the model is fitted.

```python
print(model.coef_,model.intercept_)
```

## Supervised classification of Iris with decision trees
Decision trees are a type of supervised machine learning algorithm used for both classification and regression tasks. They work by partitioning the feature space into distinct regions, aiming to separate different classes or predict numerical values based on input features. The model structure mimics a tree, where each internal node represents a decision based on one of the input features, each branch signifies an outcome of that decision, and each leaf node holds a class label or a continuous value as the final output.

A decision tree is a flowchart-like tree structure where internal nodes represent features (or attributes), branches represent decisions (or tests) on these features, and leaf nodes represent outcomes (or labels). The root of the tree is the best predictor chosen based on a given criterion. The primary mathematical formulations behind decision trees involve criteria to split data at each node. For classification tasks, common criteria include the Gini impurity, entropy, and information gain. For regression tasks, decision trees often use the variance reduction as the criterion. The goal during the tree-building process is to choose splits that best segregate the data into the desired classes (for classification) or minimize the variance of the target variable (for regression). The tree grows by recursively finding the best splits until a stopping condition is met, such as a predefined tree depth or a minimum number of samples per leaf.

```python
# look how slow pairplot is for a dataset of size (150,5)... so small! 
# at the end of the day, its very hard to use it for real world data 
sns.pairplot(iris, hue='species', height=1.5);
```

Preparing the data for training. We have to separe X (the train) from y (the target). We thus drop the class label from the DataFrame and use it as class labels. 

```python
X = iris.drop('species', axis=1)
X.shape
```

```python
y = iris['species']
y.shape
```

We then split the dataset, we want to have some data for training and some for testing!

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=1, train_size = 0.8)
print(Xtrain.shape, Xtest.shape, ytrain.shape, ytest.shape)
```

And now we fit the model class to the data

```python
#import the model class
from sklearn.tree import DecisionTreeClassifier
#instantiate the model
model = DecisionTreeClassifier(criterion = 'entropy')
# fit the model to data
model.fit(Xtrain, ytrain)      
```

And predict the data

```python
ytrain_model = model.predict(Xtrain)    
```

We can use the ``accuracy_score`` utility to see the fraction of predicted training set labels that match their true value.

```python
from sklearn.metrics import accuracy_score
accuracy_train = accuracy_score(ytrain, ytrain_model)
print("The accuracy on training set is {0:.2f}%".format(accuracy_train * 100))
```

```python
ytest_model = model.predict(Xtest)                  # 4. predict on new data
accuracy_test = accuracy_score(ytest, ytest_model)
print("The accuracy on test set is {0:.2f}%".format(accuracy_test * 100))
```

### Appendix, if we got here we are fast!

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train_int = label_encoder.fit_transform(ytrain)
y_train_cat = keras.utils.to_categorical(y_train_int, num_classes=3)

# Define a simple Sequential model
model = keras.Sequential()

# Add a dense layer with 32 units and a ReLU activation
model.add(layers.Dense(32, activation='relu', input_shape=(4,)))

# Add another dense layer
model.add(layers.Dense(32, activation='relu'))

# Add the output layer with one neuron and a sigmoid activation for binary classification
model.add(layers.Dense(3, activation='sigmoid'))

# Compile the model with a binary cross-entropy loss, adam optimizer, and accuracy as the metric
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

# Train the model for 10 epochs
model.fit(Xtrain, y_train_cat, epochs=10, batch_size=1)
```

```python
transformed_labels = label_encoder.fit_transform(ytest)
transformed_labels = keras.utils.to_categorical(transformed_labels, num_classes=3)
```

```python
ytest_model = model.predict(Xtest) 
max_indices = np.argmax(ytest_model, axis=1)
result = np.zeros(ytest_model.shape)
result[np.arange(result.shape[0]), max_indices] = 1

accuracy_test = accuracy_score(result, transformed_labels)
print("The accuracy on test set is {0:.2f}%".format(accuracy_test * 100))
```
